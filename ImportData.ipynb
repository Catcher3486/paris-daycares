```python
import pandas as pd 
import numpy as np
from bs4 import BeautifulSoup 
import requests 
# import lxml 

### #skipping the HTTP call part for now & using the downloaded HTML file#
## NB: when uncommenting check the utf encoding 

#url= "https://allocreche.fr/75/paris/micro-creche"
#url2 = "https://monenfant.fr/que-recherchez-vous?structure=EAJE&location=Paris,%20%C3%8Ele-de-France,%20FRA&x=2.35245810663&y=48.856413872242&radius=2000"
# r = requests.get(url, headers={"User-Agent": "Mozilla/5.0"})
# r.encoding = "utf-8"  
#soup = BeautifulSoup(r.text, "html.parser")


with open("src/Allocreche/MicroCrechesParis.html", encoding="utf-8") as f:
    soup = BeautifulSoup(f.read(), "html.parser")

#soup = BeautifulSoup(filehandle,"html.parser")

allocreche = soup.select('div[itemtype="http://schema.org/LocalBusiness"]')
allocreche.encoding = "utf-8"
   
def safe_text(el):
    return el.get_text(strip=True) if el else ""

for b in soup.select('div[itemtype="http://schema.org/LocalBusiness"]'):
    name   = safe_text(b.select_one('[itemprop="name"]'))
    street = safe_text(b.select_one('[itemprop="streetAddress"]'))
    postal = safe_text(b.select_one('[itemprop="postalCode"]'))
    city   = safe_text(b.select_one('[itemprop="addressLocality"]'))

    # optional: skip empty cards
    if not name:
        continue
    print(f"{name}, {street}, {postal} {city}")
```

```python
with open("src/MonEnfant/MonEnfant.html", encoding="utf-8") as f:
    soup = BeautifulSoup(f.read(), "html.parser")


results = []

for card in soup.select("div.structure-card"):
    name = safe_text(card.select_one("h4.title-content p"))
    address = safe_text(card.select_one("span.address-content"))

    # phone: first contact-txt if present
    phone = ""
    contact_spans = card.select("div.contact-card span.contact-txt")
    if contact_spans:
        phone = safe_text(contact_spans[0])

    if not name:
        continue  # skip broken cards

    results.append({
        "name": name,
        "address": address,
        "phone": phone
    })

# Preview
for r in results[:5]:
    print(r)
```
