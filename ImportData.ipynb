```python




with open("src/Allocreche/MicroCrechesParis.html", encoding="utf-8") as f:
    soup = BeautifulSoup(f.read(), "html.parser")

#soup = BeautifulSoup(filehandle,"html.parser")

allocreche = soup.select('div[itemtype="http://schema.org/LocalBusiness"]')
allocreche.encoding = "utf-8"
   
def safe_text(el):
    return el.get_text(strip=True) if el else ""

rows_ac = []

for b in soup.select('div[itemtype="http://schema.org/LocalBusiness"]'):
    name   = safe_text(b.select_one('[itemprop="name"]'))
    street = safe_text(b.select_one('[itemprop="streetAddress"]'))
    postal = safe_text(b.select_one('[itemprop="postalCode"]'))
    city   = safe_text(b.select_one('[itemprop="addressLocality"]'))

    # optional: skip empty cards
    if not name:
        continue
    
    
    rows_ac.append({
        "name": name,
        "street": street,
        "postal": postal,
        "city": city
    })

df = pd.DataFrame(rows_ac)
df
```
